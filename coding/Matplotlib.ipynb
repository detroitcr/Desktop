```python
import matplotlib.pyplot as plt
# from matplotlib import pyplot as plt
import numpy as np


%matplotlib inline  
```

```python
plt.plot([1,6,5,2],[100,250,300,200])
plt.show()
```

```python
x = np.arange(1 , 10)
y = x ** 2
plt.title('Squares')
plt.ylabel('profits')
plt.xlabel('output')
plt.plot(x,y,'r--')
plt.grid()
plt.text(5.5,30,'CR')

plt.xlim(-1,15)
plt.ylim(2,50)
plt.show
```

```python
x = np.arange(-10,20)
y = x**2
plt.plot(x,y,'r--',label = 'square' , linewidth=2)

x2 = np.arange(-10,15)
y2 = x2 + 50
plt.plot(x2 , y2, label = 'linear',linewidth=5)
plt.legend()

plt.show()
```

```python
plt.plot(x,y,'r--',x2,y2,'y--')
plt.show()
```
