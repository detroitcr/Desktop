```python
weather= pd.read_csv('data/weather.csv')
weather.head()
```

```python
weather_melt=pd.melt(weather,id_vars=['id','year','month','element'],
       var_name='day',
        value_name='temp')
weather_melt.head()
```

```python
weather_melt.pivot_table(index=['id','year','month','day',],
                        columns = 'element',
                        values = 'temp')
weather_melt.head()
```

```python
weather_tidy=weather_melt.pivot_table(index=['id','year','month','day',],
                        columns = 'element',
                        values = 'temp', dropna=False).reset_index()
weather_tidy.head()
```

```python
weather_tidy.to_csv('output/weather_tidy.csv',sep=',')
```
